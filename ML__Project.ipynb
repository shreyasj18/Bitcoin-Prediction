```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

Reading csv file of bitcoin dataset

```python
df=pd.read_csv('/content/sample_data/Bitcoin_dataset.csv',index_col=0)
df.head()
```

Calculating how many each column has nan values using isnull() and sum() function

```python
df.isnull().sum()
```

droping unnecessary columns from the dataset

```python
new_df=df.drop('fee [USD]',axis='columns')
new_df=new_df.drop('btc search trends',axis='columns')
new_df=new_df.iloc[:,:-1]
new_df.head(25)
#new_df.isnull().sum()
```

Processing and clearing the dataset . fill the nan values with average of the respective columns

```python
mean_value=new_df['Gold price[USD]'].mean()
new_df['Gold price[USD]'].fillna(value=mean_value,inplace=True)
mean_value=new_df['SP500 close index'].mean()
new_df['SP500 close index'].fillna(value=mean_value,inplace=True)
mean_value=new_df['Oil WTI price[USD]'].mean()
new_df['Oil WTI price[USD]'].fillna(value=mean_value,inplace=True)
#new_df=new_df.loc[1:364,['BTC_price','n-transactions ','btc search trends','Gold price[USD] ','SP500 close index ','Oil WTI price[USD]','M2(Not seasonally adjusted)[1e+09 USD]']]
new_df.isnull().sum()
new_df.head(25)
```

spliting dataset into target and inputs values

```python
inputs=new_df.drop('BTC_price',axis='columns')
target=new_df.BTC_price
target=target.astype(int)
# inputs.isnull().sum()
# target.isnull().sum()
```

spliting test train data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(inputs,target, test_size = 0.30,random_state=32)
```

import logistic regression model from sklearn

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()  # Create an instance of the LogisticRegression
```

```python
#training the model on the train dataset
```

```python
model.fit(X_train,Y_train)
```

```python
#calculating accuracy of the model using test data
```

```python
model_accuracy = model.score(X_test,Y_test)
print('Model Accuracy: ',round(model_accuracy*100,3))
```
